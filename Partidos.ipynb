```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("partido.csv")
```

```python
df['partido'].unique()
```

```python
plt.figure(figsize=(40,20))
plt.subplot2grid((2,3), (0,0))
df.partido.value_counts().plot(kind='bar')
plt.title('Partidos ganadores')

plt.subplot2grid((2,3), (0,1))
df.partido.value_counts(normalize=True).plot(kind='bar')
plt.title('Partidos ganadores %')

plt.show()
```

```python
plt.figure()
plt.title("Partidos Ganadores", fontsize = 30)
df.partido.value_counts().plot(kind='pie')
plt.show()
```
